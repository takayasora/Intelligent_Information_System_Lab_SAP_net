```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd


def NetworkX(df):
    G = nx.DiGraph()

    # ノードを追加
    nodes = df.columns
    for node in nodes:
        G.add_node(node)

    # エッジを追加
    for i in range(len(nodes)):
        for j in range(len(nodes)):
            if df.iloc[i, j] > 0:  # 0より大きい値のエッジを追加
                G.add_edge(nodes[i], nodes[j], weight=df.iloc[i, j])  # エッジの重みを設定

    # エッジの重みを取得
    edge_weights = nx.get_edge_attributes(G, 'weight')

    # ネットワーク図を描画
    pos = nx.spring_layout(G)  # レイアウトを設定
    nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_color='black', arrowsize=20)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights)  # エッジ上の値を表示
    plt.title("Network Graph")
    plt.show()


# データを定義
data = {
    'Adata_col': [0.0, 0.2, 0.2],
    'Bdata_col': [0.2, 0.0, 0.1],
    'Cdata_col': [0.2, 0.1, 0.0]
}

# データフレームを作成
df = pd.DataFrame(data, index=['Adata_row', 'Bdata_row', 'Cdata_row'])

# データの表示
print(df)

# データの可視化
NetworkX(df)
```
